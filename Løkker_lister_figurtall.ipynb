#### for-løkker og tallfølger
Når vi vil at et program skal gjøre samme oppgave flere ganger lager vi løkker. Dersom vi på forhånd vet hvor mange ganger vi ønsker løkken skal kjøres bruker vi en for-løkke.
```python
antall=int(input("hvor mange tall?"))

for n in range(1,antall+1):  # hvorfor +1 tror du?
    print(n)
```

### while løkker:
Når man på forhånd ikke vet hvor mange ganger man trenger å kjøre løkken kan man lage en while-løkke.
Denne løkken går *så lenge* et vilkår er sant. while løkker trenger at variablen som skal sjekkes er gitt på forhånd og at variabelen endres inni løkken slik at løkken til slutt stopper.

<div class="alert alert-block alert-warning">
<b>Eksempel: </b>Lag en løkke som så lenge et tall er positivt, skriver ut tallet, trekker fra 2 og gjentar.  </div>

```python
tall = 15  #variabelen talldefineres og gies en verdi FØR løkken 

while tall>0: # så lenge variabelen er større enn null, kjører løkken.
    print(tall)
    tall -= 2 #python for: ta det opprinnelige tallet, trekk fra to og oppdater variabelen 'tall' 
                # Hva tror du += gjør?
```

### Lister
Ofte er det lurt å ta vare på resultatene vi lager i en liste. Her lærer du å lage og fylle en liste.
<div class="alert alert-block alert-warning">
<b>Eksempel: </b>Lag en tom liste med navnet 'minliste', legg inn tallene 1,2,3, ... ,10 og 
skriv resultatet til skjerm  </div>

```python
antall = 10
minliste =[] # firkant paranteser for liste, denne er tom

for i in range(1,antall + 1):
    minliste.append(i)  # På minliste, bruk listemetoden append. 
                        # Den legger til det vi har i parantesen til slutt i listen
print(minliste) # skriver ut resultatet
```

I koden under har vi lekt litt med en liste og en teller. Kanskje forstår du hvordan du kan korigere svaret ditt over dersom while løkken kjører en ekstra gang.

```python
# av og til går whileløkker en gang mer en man ønsker. Kan koden under være til hjelp?

# la oss si at dette er resultatet etter en løkke som har kjørt en gang for mye:
n = 5
liste = [0,1,2,3,4]
print('Verdier før endring')
print(n)
print(liste)

# etter at vi sjekker resultatet, kan vi korrigerer litt:
n = n-1     # Ruller telleren tilbake ett hakk 
liste.pop() # pop ut det siste elementet 
print('\nVerdier etter endring') #\n betyr "ny linje" slik at det blir luft i det som printes ut.
print(n)
print(liste)
```
