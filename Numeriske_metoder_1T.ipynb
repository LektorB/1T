## kp 4.5 Halveringsmetoden
Løse likninger numerisk

```python
#lager en funksjon av likingen

def f(x):
    return x**2-5

nedre = -4 # nullpunktet er mellom disse to grensene 
øvre = -1

grense = 0.0001

while øvre - nedre > grense:
    midt = (øvre + nedre)/2 # regner ut midtpunktet
    
    if f(øvre)*f(midt) > 0: # midtpuktet og øvregrense på samme side av nullpuktet
        øvre = midt
    else:
        nedre = midt
        
print(f'vi finner løsningen x = {midt:.4f}') # format 4 desimaler, flyttal
```

## kp 5.6 numerisk derivasjon

$$ f'(x) = \lim_{\Delta x\to 0}\frac{f(x+\Delta x)-f(x)}{\Delta x}\approx \frac{f(x+h)-f(x)}{h}$$

der vi velger et lite tall for h

```python
# importere biblioteke vi trenger
import matplotlib.pyplot as plt
import numpy as np


start = -5
stopp = 8
hopp = 0.1

# definere funksjonen vi skal tegne
def f(x):
    return 2*x**2-4*x+1

def derivert(x, h=0.001):
    return (f(x+h)-f(x))/h


# lister/arrayer med x og yverdier
xverdier = np.arange(start,stopp,hopp)
yverdier = f(xverdier)
der_verdier = derivert(xverdier)

# plotte dataene
plt.plot(xverdier,yverdier,label = 'f(x)')
plt.plot(xverdier,der_verdier,'--r',label = "f'(x)")

plt.legend()
```

```python
derivert(2,0.00000001)
```
